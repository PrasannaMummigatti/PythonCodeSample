```python
#plt.figure(figsize=(3,3))
#plt.axes(xlim=(0,100),ylim=(2,6))
df=pd.DataFrame({'a':[1,2,3,4,5,6]})
x=list(range(1,1000))
l=[]
for i in x:
    l.append(list(df.sample(3,replace=True)['a']))
y=[np.mean(x) for x in l]

#y1=[]
#y2=[]
#for n in range(len(y)):
#    for m in range(n):
#        y1.append(y[m])
#    y2.append(y1)
#    y1=[]
#yhat=[np.mean(k) for k in y2]

yhat1= [np.mean(j) for j in  [[y[i] for i in range(m)] for m in x ]]
plt.plot(x,yhat1)
plt.plot(x,[3.5]*len(x))
np.mean(yhat1)
#plt.plot(x,yhat)
```
