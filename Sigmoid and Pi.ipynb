```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))


def sigmoid2(x1,x2,x3):
    return 1/(1+np.exp(-(0.3+3*x1+0.001*x2+0.0045*x3)))
```

```python
plt.figure(figsize=(30,20))
plt.axis(ymin=-0.2,ymax=1.2)
x=np.linspace(-10,10,200)
x1=np.linspace(-10,10,200)
x2=np.linspace(-10,10,200)
x3=np.linspace(-10,10,200)

y=sigmoid(x)
y2=sigmoid2(x1,x2,x3)


plt.plot(x,y)
plt.plot(x,y2)

plt.grid()


plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=list(range(5000))
def simulate(x):
    incr=0
    y=[np.sqrt(np.random.rand()**2+np.random.rand()**2) for i in list(range(x))]
    for j in y:
        if j<1:
            incr=incr+1
            #print(j)
    return 4*(incr/x)
    


# function call    
    
y=[simulate(100) for i in x] 

np.mean(y)
yhat=[np.mean(j) for j in  [[y[i] for i in range(m)] for m in x ]]

plt.plot(x,yhat)
plt.plot(x,[np.pi]*len(x))
```
