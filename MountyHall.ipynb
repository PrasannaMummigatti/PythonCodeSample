```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
def MontyHallSimulate(x):
    Doors=[1,2,3]
    WinCountOnSwitch=0
    WinCountOnNoSwitch=0
        
    for i in range(x):
        Door_with_prize=random.sample(Doors,1)
        Door_of_playerchoice=random.sample(Doors,1)
        #print('Door_with_prize ', str(Door_with_prize), 'Door_of_playerchoice ', str(Door_of_playerchoice))

        Choices_to_OpenDoors=list(set(Doors)-(set(Door_with_prize+Door_of_playerchoice)))
        #print('Choices_to_OpenDoors',str(Choices_to_OpenDoors))

        Opened_Door=([Choices_to_OpenDoors[0]])
        #print('Opened_Door',str(Opened_Door))

        Doors_choice_toSwitch=list(set(Doors)-set(Opened_Door)-set(Door_of_playerchoice))
        #print('Doors_choice_toSwitch',str(Doors_choice_toSwitch))

        # if switch is made....
        if Doors_choice_toSwitch==Door_with_prize:
            WinCountOnSwitch=WinCountOnSwitch+1

        #if Switch is not made
        if Door_of_playerchoice==Door_with_prize:
            WinCountOnNoSwitch=WinCountOnNoSwitch+1
    return WinCountOnSwitch/x,WinCountOnNoSwitch/x
```

```python
l=[]
x=list(range(1000))
for i in x:
    [WithSwitch, WithoutSwitch]=MontyHallSimulate(100)
    l.append([WithSwitch, WithoutSwitch])


WithSwitch=[np.mean([l[i][0] for i in range(m)]) for m in range(1,len(l)+1)]
WithoutSwitch=[np.mean([l[i][1] for i in range(m)]) for m in range(1,len(l)+1)]

plt.plot(x,WithSwitch,label="WithSwitchWinProbability")
plt.plot(x,WithoutSwitch,label="WithoutSwitchWinProbability")
plt.legend()
    
```
