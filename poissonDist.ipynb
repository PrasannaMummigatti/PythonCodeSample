```python
def fun(la,K):
    return [((la**k)*np.exp(-la))/np.math.factorial(k) for k in K]
```

```python
la=1
k=np.linspace(0,20,21)
```

```python
a=fun(1,list(k))
b=fun(2,list(k))
c=fun(3,list(k))
d=fun(4,list(k))
e=fun(11,list(k))
plt.plot(k,a)
plt.plot(k,b)
plt.plot(k,c)
plt.plot(k,d)
plt.plot(k,e)
```

```python
plt.figure(figsize=(10,5))
[plt.plot(k,[fun(i,list(k)) for i in range(0,10)][j]) for j in range(10)]
```

```python
plt.figure(figsize=(20,5))
m=[plt.plot(k,[fun(i,list(k)) for i in list(np.linspace(1,7,100))][j]) for j in range(100)]
```
