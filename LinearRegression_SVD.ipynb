```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize']=[8,8]
plt.rcParams.update({'font.size':18})
```

```python
x = 3; # True slope
a = np.arange(-2,2,.525);
a=a.reshape(-1,1)
b = a*x + 1*np.random.randn(*a.shape); # Add noise
plt.plot(a,x*a,Color='k',LineWidth=2,label='True Line') # True relationshiphold on, plot(a,b,rx) # Noisy measurements
plt.plot(a,b,'x',Color='r',Markersize=10,label='Noisy Data')

U,S,VT = np.linalg.svd(a,full_matrices=False);
xtilde = VT.T@np.linalg.inv(np.diag(S))@ U.T @ b  # Least-square fit
xtilde2 = np.linalg.pinv(a) @ b  # Least-square fit

plt.plot(a, xtilde*a ,'--',Color='r',Markersize=10,label='Regression Line') # Plot fit
plt.plot(a, xtilde2*a ,'--',Color='y',Markersize=10,label='Regression Line2') # Plot fit

plt.xlabel('a')
plt.ylabel('b')
plt.grid(linestyle='--')
plt.legend()
plt.show()
```
